```python
from random import randint

gesture = ["rock", "paper", "scissors"]
n_rounds = 0 
max_rounds = 5
rounds_to_win = (max_rounds + 1)//2
tie = 0
cpu_score = 0
player_score = 0

def player_choice():
    player = input("\nChoose Rock, Paper, Scissors?: ")
    player = player.lower()
    while player not in gesture:
        player = input("please choose rock paper scissors!")
        player = player.lower()
        
    return player
    

def computer_choice():
    computer = gesture[randint(0,2)]
    return computer
    
def result(player, computer):
    if player == computer:
        return 0
    elif player == "rock":
        if computer == "paper":
            return 1
        else:
            return 2
    elif player == "paper":
        if computer == "scissors":
            return 1
        else:
            return 2
    elif player == "scissors":
        if computer == "rock":
            return 1
        else:
            return 2

def announce(player, computer, outcome):
    global tie, cpu_score, player_score
    print("Player has chosen ", player)
    print(f"Computer has chosen {computer}")
    if outcome==0:
        tie+=1
        print("Tie")
    elif outcome==1:
        cpu_score+=1
        print("Computer won!")
    elif outcome==2:
        player_score+=1
        print("You won!")
        
    
def score(player_score, cpu_score):
    print("Player wins:", player_score)
    print("CPU wins:", cpu_score)
    print("tie:", tie)
    if player_score > cpu_score:
        print("you won!")
    else:
        print("you lost!")
    
    
while n_rounds != max_rounds and cpu_score < rounds_to_win and player_score < rounds_to_win:
    n_rounds += 1
    print(f"Round: {n_rounds}")
    player = player_choice()
    cpu = computer_choice()
    outcome = result(player, cpu)
    announce(player, cpu, outcome)
    
score(player_score, cpu_score)  
```
